```python
#Dictionary

def dictionary():
    x=65;
    bandera=0;
    d=[]
    for i in range(61):
        if (i<26):
            d.append(x)
            x+=1
        if(i>=26 and i<52 and bandera==0):
            x=97
            bandera=1;
        if(i>=26 and i<52):
            d.append(x)
            x+=1
        if(i==52):
            d.append(34)
        if(i==53):
            d.append(40)
        if(i==54):
            d.append(41)
        if(i==55):
            d.append(63)
        if(i==56):
            d.append(39)
        if(i==57):
            d.append(58)
        if(i==58):
            d.append(59)
        if(i==59):
            d.append(44)
        if(i==60):
            d.append(46)
    return d
```

```python
#Encrypt
def encrypt(text, k):
    result = ""
    d=dictionary()
    l=len(k)
    j=0
    for i in range(len(text)):
        char = text[i]
        keyword = k[j]
        if(ord(char)!=32):
            m = d.index(ord(char))
            n = d.index(ord(keyword))
            r = (m+n) % 61
            e = d[r]
            result += chr(e)
            j += 1
            if(j==l):
                j=0
    return result
```
